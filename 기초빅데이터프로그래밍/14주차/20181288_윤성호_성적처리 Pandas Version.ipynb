```python
## stuDta.csv가 현재 파이썬 파일과 같은 디렉토리에 있다고 가정하고 프로그램을 작성하였습니다.
## 한 학기동안 감사했습니다!

import pandas as pd
import numpy as np

# csv확장자 파일에 저장되어 있는 학생정보를 pandas를 이용해 받아온다.
stuData = pd.read_csv('stuData.csv', index_col=0)

# Computer 과목의 점수를 dataFrame에 추가해준다.
stuData['Com'] = [78, 99, 90, 45, 80, 75]

# 점수의 평균을 구하여 dataFrame에 추가해준다.
stuData["평균"] = (stuData["국어"] + stuData["영어"] + stuData["수학"] + stuData["Com"])/4

# 그래프를 그리기 위해 미리 학생들의 평균을 구해놓는다.
stuAvg = stuData["평균"]

## 각 각과목의 평균을 추가해주기 위해 구해서 저장한다.
korAvg = float(f"{stuData['국어'].mean():.2f}")
engAvg = float(f"{stuData['영어'].mean():.2f}")
matAvg = float(f"{stuData['수학'].mean():.2f}")
comAvg = float(f"{stuData['Com'].mean():.2f}")
avgAvg = float(f"{stuData['평균'].mean():.2f}")

# "평균" 이르는 이름의 행에 정보를 저장해준다.
stuData.loc["평균"] = [np.nan, korAvg, engAvg, matAvg, comAvg, avgAvg]

# dataFrame을 출력해준다.
stuData
```

```python
import matplotlib.pyplot as plt

# 전체 figure의 사이즈에 접근하기 위해 변수로 설정하고 그 크기를 설정하였다.
fig = plt.figure(figsize=(50,20))

############################### 과목별 평균 그래프 ###############################

subjectAvg = [korAvg, engAvg, matAvg, comAvg]
SUBAVG = fig.add_subplot(2,7,3)
SUBAVG.set_title("Subject Average", size = 30)
SUBAVG.set_xlabel("Subjects", size = 20)
SUBAVG.set_ylabel("Average score", size = 20)
SUBAVG.tick_params(axis='both', which='major', labelsize=20)

SUBAVG.plot(["kor", "eng", "math", "Com"],subjectAvg, 'bo-')

########################### End of 과목별 평균 그래프 ############################



############################### 학생별 평균 그래프 ###############################

STUAVG = fig.add_subplot(2,7,5)
STUAVG.set_title("Average score of students", size = 30)
STUAVG.set_xlabel("Students", size = 20)
STUAVG.set_ylabel("Average score", size = 20)
STUAVG.tick_params(axis='both', which='major', labelsize=20)


STUAVG.plot(["hg1", "hg2", "hg3", "hg4", "hg5", "hg6"], stuAvg, 'bo-')

########################### End of 과목별 평균 그래프 ############################


############################### 각 학생의 점수 그래프 #############################

Student = fig.add_subplot(2,7,11)

hongil1 = stuData.loc[2201]["국어":"Com"]
hongil2 = stuData.loc[2202]["국어":"Com"]
hongil3 = stuData.loc[2203]["국어":"Com"]
hongil4 = stuData.loc[2204]["국어":"Com"]
hongil5 = stuData.loc[2205]["국어":"Com"]
hongil6 = stuData.loc[2206]["국어":"Com"]

Student.plot(["kor", "eng", "math", "Com"], hongil1, label="hg1", marker="o")
Student.plot(["kor", "eng", "math", "Com"], hongil2, label="hg2", marker="o")
Student.plot(["kor", "eng", "math", "Com"], hongil3, label="hg3", marker="o")
Student.plot(["kor", "eng", "math", "Com"], hongil4, label="hg4", marker="o")
Student.plot(["kor", "eng", "math", "Com"], hongil5, label="hg5", marker="o")
Student.plot(["kor", "eng", "math", "Com"], hongil6, label="hg6", marker="o")

Student.legend(fontsize=20)
Student.set_title("Scores of each student", size = 30)
Student.set_xlabel("Students", size = 20)
Student.set_ylabel("Score", size = 20)
Student.tick_params(axis='both', which='major', labelsize=20)

######################### End of 각 학생의 점수 그래프 ###########################
```
