```python
# 실습 4

import matplotlib.pyplot as plt
import numpy as np

## 원주율값을 PI라는 변수에 저장해놓는다.
PI = np.pi

## linspace 함수를 이용하여 -pi, pi 사이에 등간격으로 256개의 값을 생성한다.
X = np.linspace(-PI, PI, 256)

## 그렇게 생성한 X값을 cos함수에 대입하여 Y 배열을 생성한다.
Y = np.cos(X)

## XKCD 형식의 그래프를 그리기 위한 과정이다.
with plt.xkcd():
    plt.title("XKCD style plot!!!") ## title을 써준다.
    plt.plot(X, Y) ## 그래프를 그려준다.
    plt.scatter([PI*2/3], [np.cos(PI*2/3)], color="r", linewidth=4) ## 점을 빨간색으로 표시해준다.
    
    ## -0.5 Here! 이라는 문구가 해당 점을 화살표로 가리키도록 한다.
    plt.annotate('-0.5 Here!', xy=(PI*2/3, np.cos(PI*2/3)), xycoords='data', xytext=(-85,-50), textcoords="offset points", arrowprops=dict(arrowstyle="->"))
```

```python
# 과제 

import matplotlib.pyplot as plt
import numpy as np

## XKCD 형식으로 그래프를 그리기 위한 과정이다.
with plt.xkcd():
    plt.xlabel("time") ## xlabel을 써준다.
    plt.ylabel("my overall health") ## ylabel을 써준다.
    
    ## 현재 axes의 오른쪽 부분을 지워주기위해 색으로 none값을 주었다.
    plt.gca().spines['right'].set_color('none')
    
    ## 현재 axes의 오른쪽 부분을 지워주기위해 색으로 none값을 주었다.
    plt.gca().spines['top'].set_color('none')
    
    ## 현재 axes의 y값을 표시하는 범위를 0부터 53으로 제한한다.
    plt.gca().set_ylim([0,53])
    
    plt.xticks([]) ## x값들을 지워준다.
    plt.yticks([]) ## y값들을 지워준다.
    
    ## 함수의 Y값을 생성하는 부분이다.
    Y1 = np.linspace(40,40,600)
    Y2 = np.arange(40,0,-0.1)
    Y1 = np.append(Y1, Y2)
    
    plt.plot(Y1) # 함수를 그려준다.
    plt.annotate("THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED", xy=(600,40), arrowprops=dict(arrowstyle="->"), xytext=(80,25))
    
    plt.gcf().text(0.5, -0.01,'"Stove Ownership" from xkcd by Randall Monroe', ha='center')
```
