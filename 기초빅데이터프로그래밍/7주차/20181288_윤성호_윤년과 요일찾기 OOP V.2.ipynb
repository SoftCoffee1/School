```python
## 데코레이터 

import time

## checkNowTime은 객체 생성시각을 알려주는 데코레이터로서 활용하기 위해 리턴값을 함수로 주었다.
def checkNowTime(func):
    
    ## 인자의 유무에 관계없이 처리를 할 수 있도록 *args, **kwargs를 이용하였다.
    def new_func(*args, **kwargs):
        
        ## 현재 실행시작 날짜를 주어진형식으로 출력한다.
        ## args[0]은 현재 객체를 가리키는 self를 저장하고 있다.
        print()
        print(f"\n{type(args[0]).__name__} 클래스의 객체 생성 시각:",time.strftime('[%Y-%m-%d %H:%M]', time.localtime(time.time())))
        
        ## 데코레이터가 적용될 함수를 동작시키는 부분이다.
        func(*args, **kwargs)
        
    ## 리턴값으로 함수를 준다.
    return new_func
    

    
## checkRunningTime은 함수의 실행시간을 측정해준다.
def checkRunningTime(func):
    
    def new_func(*args, **kwargs):
        
        ## 시작시각 저장하기
        start_time = time.time()
        
        ## 함수 실행
        func(*args, **kwargs)
        
        ## 종료시각 저장하기
        end_time = time.time()
        
        print("실행시간은 : ", end_time - start_time)
        print()
        print()
        
    return new_func
    
```

```python
## EverythingAboutDate 클래스 정의    
class EverythingAboutDate:
    
    ## 생성된 객체의 개수
    _count = 0
    
    ## 생성자 : year, month, day 변수를 생성한다. 그리고 입력날짜의 유효성 여부를 담는 isValidDate, 윤년인지 아닌지를 담는 변수 isLeap을 생성한다.
    @checkNowTime
    def __init__(self, year, month, day):
        self._year = year
        self._month = month
        self._day = day
        EverythingAboutDate._count += 1
        print(f"현재 객체의 수는 {EverythingAboutDate._count}개이다.")
        print("\n\n\n")
        
    
    def __del__(self):
        EverythingAboutDate._count -= 1
        print("객체가 삭제되었습니다.")
        print(f"현재 객체의 수는 {EverythingAboutDate._count}개이다.")
        print("\n\n\n")
        
        
    
    
    ## 유효한 날짜가 입력되면 요일의 이름을 반환하주는 매서드
    def get_name(self):
    
        ## 유효한 입력이 아니면 에러메세지를 반환한다.
        if not DayUTILS.is_valid_date(self._year, self._month, self._day):
            return "입력하신 날짜는 존재하지 않습니다."
        

        year = int(self._year)
        month = int(self._month)
        day = int(self._day)
        
        ## 윤년이면 
        if DayUTILS.is_leap(year):
            
            ## 입력한 년도가 윤년이라는 문장을 출력해야 한다.
            isLeap = "\n입력하신 "+str(year)+"은 윤년입니다."
        
        ## 윤년이 아니면
        else:
            isLeap = ""
            
        
        ## 년월일 형식의 날짜 데이터를 일수로 변환한다.
        convertedDay = DayUTILS.date2day(year, month, day)
            
        
        ## 그 값을 7로 나누면 요일이 나온다.
        day_name = convertedDay % 7


        ## 1년 1월 1일이 월요일이므로, 인덱스 1값을 월요일로 주었다.
        DAYNAME = ["일", "월", "화", "수", "목", "금", "토"]
        
        ## 현재 요일과, 현재 년도의 윤년여부가 저장되어있는 isLeap을 같이 문자열로 묶어 리턴값으로 준다.
        return self._year+"년 "+self._month+"월 "+self._day+"일은 "+DAYNAME[day_name] + "요일입니다." + isLeap
    
    
    
    ##########################################################################################################
    ##  < 특정 년도 사이의 윤년의 개수 구하기 >                                                                     ##
    ## 1. leap_count(*years) : 년도를 입력받아 그 사이의 윤년 개수를 반환한다.                                         ##
    ##########################################################################################################
    
    @checkRunningTime
    def leap_count(self, *years):
        
        ## 윤년개수를 저장하는 변수
        count = 0
        
        ## 입력값이 없다면 1년부터 올해까지의 년도 중 윤년의 개수를 반환한다.
        if len(years) == 0:
            start = 1
            end = int(self._year)
            
        ## 입력값이 하나라면 1년부터 입력년도까지의 년도 중 윤년의 개수르 반환한다.
        elif len(years) == 1:
            start = 1
            end = years[0]
            
        ## 입력값이 하나라면 1년부터 입력년도까지의 년도 중 윤년의 개수르 반환한다. 두 수의 입력순서는 상관이 없다.
        elif len(years) == 2:
            start = min(years[0], years[1])
            end = max(years[0], years[1])
            
        else:
            print("입력 날짜가 너무 많습니다.")
            return
        
        for year in range(start, end+1):
            if DayUTILS.is_leap(year):
                count += 1
                
        
        print(f"{start}년부터 {end}년 사이의 윤년의 개수는 {count}개입니다.")
    
    ###########################     End of < 특정 년도 사이의 윤년의 개수 구하기 >    ################################
    
    
    ##########################################################################################################
    ##  < 현재 날짜로부터 며칠 후의 날짜 구하기 >                                                                    ##
    ## 1. whatDay(num) : 현재 날짜에서 num일 후의 날짜를 반환한다. num은 양의 정수, 음의정수 모두 가능하다.                   ##
    ##########################################################################################################
    
    def whatDay(self,num):
        
        year = int(self._year)
        month = int(self._month)
        day = int(self._day)
        
        ## 현재 날짜를 일수롤 변환한다.
        convertedDay = DayUTILS.date2day(year, month, day)
        
        ## 새로운 날짜를 일수형식으로 구한다.
        newDay = convertedDay + num
        
        ## 새로운 날짜를 일수에서 년월일로 변환한다.
        year, month, day = DayUTILS.day2date(newDay)
        
        if num > 0:
            print(f"{self._year}년 {self._month}월 {self._day}일에서 {num}일 후는 {year}년 {month}월 {day}일이다.")
            
        elif num < 0:
            print(f"{self._year}년 {self._month}월 {self._day}일에서 {-num}일 전은 {year}년 {month}월 {day}일이다.")
        
    
    ###########################     End of < 현재 날짜로부터 며칠 후의 날짜 구하기 >    ###############################
            
            
    
    ##########################################################################################################
    ##  < +, - 연산자 오버라이딩 >                                                                              ##
    ## 1. __add__(num) : num일 후의 날짜정보를 알려주도록 오버라이딩하였다.                                            ##
    ## 2. __sub__(num) : num일 전의 날짜정보를 알려주도록 오버라이딩하였다.                                            ##
    ##########################################################################################################
    
    def __add__(self, num):
        
        year = int(self._year)
        month = int(self._month)
        day = int(self._day)
        
        ## 현재 날짜를 일수롤 변환한다.
        convertedDay = DayUTILS.date2day(year, month, day)
        
        ## 새로운 날짜를 일수형식으로 구한다.
        newDay = convertedDay + num
        
        ## 새로운 날짜를 일수에서 년월일로 변환한다.
        year, month, day = DayUTILS.day2date(newDay)
        
        print(f"{self._year}년 {self._month}월 {self._day}일에서 {num}일 후는 {year}년 {month}월 {day}일이다.")
    
    def __sub__(self, num):
        
        year = int(self._year)
        month = int(self._month)
        day = int(self._day)
        
        ## 현재 날짜를 일수롤 변환한다.
        convertedDay = DayUTILS.date2day(year, month, day)
        
        ## 새로운 날짜를 일수형식으로 구한다.
        newDay = convertedDay - num
        
        ## 새로운 날짜를 일수에서 년월일로 변환한다.
        year, month, day = DayUTILS.day2date(newDay)
        
        print(f"{self._year}년 {self._month}월 {self._day}일에서 {num}일 전은 {year}년 {month}월 {day}일이다.")
         
        
    ################################     End of < +, - 연산자 오버라이딩 >     ####################################
    
    
    ##########################################################################################################
    ##  < 현재 날짜로부터 며칠 후의 날짜 구하기 >                                                                    ##
    ## 1. dateDelta(dateString) : 현재 날짜와 dateString 날짜의 일수 차이를 리턴하는 매서드                            ##
    ##########################################################################################################
    
    
    def dateDelta(self, dateString):
        
        ## 입력 날짜의 년, 월, 일
        year1, month1, day1 = map(int, dateString.split(","))
        
        ## 현재 객체의 년, 월, 일
        year, month, day = int(self._year), int(self._month), int(self.day)
        
        ## 입력날짜의 년월일을 일수로 변환
        inputDay = DayUTILS.date2day(year1, month1, day1)
        
        ## 현재 객체의 날짜의 년월일을 일수로 변환
        nowDay = DayUTILS.date2day(year, month, day)
        
        ## 일수의 차이를 계산
        datedelta = abs(inputDay - nowDay)
        
        print(f"현재 날짜({year}년 {month}월 {day}일)와 {year1}년 {month1}월 {day1}일은 {datedelta}일 차이가 난다.")
        
    ############################     End of < 현재 날짜로부터 며칠 후의 날짜 구하기 >    ###############################
    
    
    ##########################################################################################################
    ##  < property class를 이용하여 매서드를 속성값에 접근하듯이 접근 >                                                ##
    ## 1. get_year() : year을 반환하는 매서드                                                                   ##
    ## 2. set_year() : year을 변경하는 매서드                                                                   ##
    ## 3. get_month() : month를 반환하는 매서드                                                                 ##
    ## 4. get_month() : month을 반환하는 매서드                                                                 ##
    ## 5. get_day() : day을 반환하는 매서드                                                                     ##
    ## 6. set_day() : day을 변경하는 매서드                                                                     ##
    ##########################################################################################################
    
    def get_year(self):
        return self._year
    
    def set_year(self, yearInput):
        self._year = yearInput
        
    def get_month(self):
        return self._month
    
    def set_month(self, monthInput):
        self._month = monthInput
        
    def get_day(self):
        return self._day
    
    def set_day(self, dayInput):
        self._day = dayInput
        
        
    ## property class 이용하기
    year  = property(get_year, set_year)
    month = property(get_month, set_month)
    day   = property(get_day, set_day)
    
    
    ##################     End of < property class를 이용하여 매서드를 속성값에 접근하듯이 접근 >   ######################
        

    ## 입력을 위한 함수
    @staticmethod
    def input_date():
    
        year = input("__년도를 입력하시오 : ")
        month = input("__월을 입력하시오 : ")
        day = input("__일을 입력하시오 : ")
    
        return year, month, day
```

```python
## DayUTILS 클래스 ##

## EverythingAboutDate 클래스에서 독립적으로 필요한 매서드들을 정적 매서드로 정의해놓은 클래스
class DayUTILS:
    
    ##########################################################################################################
    ##  < 입력데이터의 유효성 검사 >                                                                              ##
    ## 1. isValidDate(year, month, day) : 입력한 날짜가 유효한 지 체크하는 함수. 유효하면 1, 유효하지 않으면 0을 반환        ##
    ## 2. isNum(string) : string이 정수형문자로만 이루어져 있는 지 체크하는 함수. 유효하면 1, 유효하지 않으면 0을 반환          ##
    ## 3. is_valid_year(year) : year이 유효한지 체크하는 함수. 유효하면 1, 유효하지 않으면 0을 반환                       ##
    ## 4. is_valid_month(month) : month가 유효한지 체크하는 함수. 유효하면 1, 유효하지 않으면 0을 반환                    ##
    ## 5. iv_valid_day(year, month, day) : day가 유효한지 체크하는 함수. 유효하면 1, 유효하지 않으면 0을 반환.            ##
    ##########################################################################################################
    
    ## 입력 날짜의 유효성 검사 함수
    @staticmethod
    def is_valid_date(year, month, day):

        ## 년도, 월, 일이 모두 유효하면 유효한 날짜이다.
        if DayUTILS.is_valid_year(year) and DayUTILS.is_valid_month(month) and DayUTILS.is_valid_day(year, month, day):
            return True

        else:
            return False
    
    
    
    ## 정수형 입력 여부 체크를 위한 함수
    @staticmethod
    def isNum(string):
    
        ## for loop로 각각의 char를 확인하며 아스키코드상 "0"과 "9" 사이에 있는지 체크한다.
        for letter in string:
            if not ("0" <= letter <= "9"):
                return False
        
        return True

    ## 유효한 년도를 입력했는지 여부 판별 함수
    @staticmethod
    def is_valid_year(year):

        ## 정수가 아니라면 유효하지 않다.
        if not DayUTILS.isNum(year):
            return False

        ## year은 1이상의 정수만 유효하다.
        if 1 <= int(year):
            return True

        else:
            return False

    ## 유효한 달을 입력했는지 여부 판별 함수
    @staticmethod
    def is_valid_month(month):

        ## 정수가 아니라면 유효하지 않다.
        if not DayUTILS.isNum(month):
            return False

        ## month는 1이상 12이하만 유효하다.
        if 1<= int(month) <= 12:
            return True

        else:
            return False


    ## 유효한 일을 입력했는지 여부 판별 함수
    ## year과 month에 따라서 같은 day가 유효할 수도, 아닐수도 있기때문에 파라미터로 셋다 집어넣어주어야 한다.
    @staticmethod
    def is_valid_day(year, month, day):

        ## 정수가 아니라면 유효하지 않다.
        if not DayUTILS.isNum(day):
            return False

        ## 윤년일때와 아닐때는 2월의 날이 다르다. 
        ## 각 리스트의 첫번째 인덱스값은 dummy value이다.
        DaysInMonth = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        DaysInMonthLeap = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]


        ## 달이 유효하고
        if DayUTILS.is_valid_month(month):

            ## 윤년이면 
            if DayUTILS.is_leap(year):

                ## 해당 달에서 나올 수 있는 날이면 유효하다. 
                ## 달을 DaysInMonthLeap 리스트로 체크한다.
                if 1 <= int(day) <= DaysInMonthLeap[int(month)]:
                    return True

            ## 윤년이 아니면
            else:

                ## 해당 달에서 나올 수 있는 날이면 유효하다.
                ## 달을 DaysInMonth 리스트로 체크한다.
                if 1 <= int(day) <= DaysInMonth[int(month)]:
                    return True


        return False
    
    ###################################     End of < 입력데이터의 유효성 검사 >   ###################################
    
    
    
    ##########################################################################################################
    ## <윤년 검사>                                                                                            ##
    ## 1. is_leap(year) : year이 윤년인지 체크하는 함수. 윤년이면 1, 윤년이 아니면 0을 반환.                             ##
    ##########################################################################################################
    
    ## 윤년 체크 함수
    @staticmethod
    def is_leap(year):
    
        ## 계산하기 위해 정수형으로 바꿔준다.
        year = int(year)

        ## 400의 배수이면 윤년이다.
        if year % 400 == 0:
            return True

        ## 400의 배수가 아닌 수중에
        else:

            ## 100의 배수이면 윤년이 아니다.
            if year % 100 == 0:
                return False

            ## 100의 배수도 아닌 수중에
            else:

                ## 4의 배수이면 윤년이다.
                if year % 4 == 0:
                    return True

        ## 나머지는 모두 윤년이 아니다.
        return False
    
    
    #########################################     End of <윤년 검사>   #########################################
    
    
    ##########################################################################################################
    ## < 년월일 형식의 날짜를 일수로 변환하기 >                                                                      ##
    ## 1. date2day(year, month, day) : 년원일 형식의 날짜를 일수로 변환한다. 리턴값은 일수.                             ##
    ##########################################################################################################
    
    @staticmethod
    def date2day(year, month, day):
        
        ## 윤년일때와 아닐때는 2월의 날이 다르다. 
        ## 각 리스트의 첫번째 인덱스값은 dummy value이다.
        DaysInMonth = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        DaysInMonthLeap = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        
        ######### 년도를 일수로 바꾸기 #########

        ## 입력한 년은 모두 지난 것이 아니므로 년도를 일수로 바꿀때는 전년도(year-1)까지만 생각한다.
        year4 = (year-1) // 4
        year100 = (year-1) // 100
        year400 = (year-1) // 400

        leapCount = year4 - year100 + year400 ## 윤년을 세서 저장하는 변수
        ordiCount = (year-1) - leapCount      ## 윤년이 아닌 년도를 세서 저장하는 변수

        ## 윤년은 2월 29일까지 포함하므로 366일
        ## 윤년이 아니면 365일이다.
        year2Day = leapCount*366 + ordiCount*365


        ######### 월을 일수로 바꾸기 #########
        month2Day = 0

        ## 윤년이면 윤년기준 달로 보기
        if DayUTILS.is_leap(year):
            
            for m in range(month):
                month2Day += DaysInMonthLeap[m]

        ## 윤년이 아니면 윤년 아닌 기준 달로 보기
        else:
            
            for m in range(month):
                month2Day += DaysInMonth[m]


        ## 서기 1년 1월 1일부터 입력된 날까지의 날 수를 모두 더한 값
        convertedDay = year2Day + month2Day + day
        
        return convertedDay

    ##############################     End of < 년월일 형식의 날짜를 일수로 변환하기 >    #############################
    
    
    ##########################################################################################################
    ## < 일수를 년월일 형식의 날짜로 변환하기 >                                                                      ##
    ## 1. day2date(num) : 일수를 년월일의 형식으로 변환한다. 리턴값은 년, 월, 일이다.                                    ##
    ##########################################################################################################
    
    @staticmethod
    def day2date(num):
    
        ## 윤년일때와 아닐때는 2월의 날이 다르다. 
        ## 각 리스트의 첫번째 인덱스값은 dummy value이다.
        DaysInMonth = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        DaysInMonthLeap = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]


        year = 1
        month = 1
        day = 1

        while num >= 366:
            if DayUTILS.is_leap(year):
                year += 1
                num -= 366

            else:
                year += 1
                num -= 365

        ## 현재 년도가 윤년이면
        if DayUTILS.is_leap(year):

            while num >  DaysInMonthLeap[month]:
                num -= DaysInMonthLeap[month]
                month += 1

        ## 현재 년도가 윤년이 아니면 
        else:
            while num >  DaysInMonth[month]:
                num -= DaysInMonth[month]
                month += 1

        day = num


        return year, month, day

    ##############################     End of < 일수를 년월일 형식의 날짜로 변환하기 >    #############################
    
    
    
    
```
