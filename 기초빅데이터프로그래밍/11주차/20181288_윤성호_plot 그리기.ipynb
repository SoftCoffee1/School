```python
# 실습 1

import matplotlib.pyplot as plt

## bp- : 파란색선으로 그리고, 점들은 오각형을 그린다.
## lw = 5 : 선의 굵기는 5이다.
## ms = 15 : 마커의 크기는 15이다.
## mec = "g" : 마커의 테두리 색깔은 초록이다.
## mew = 5 : 마커의 테두리 굵기는 5이다.
## mfc = "r" : 마커의 내부 색깔은 빨강이다.

plt.plot([1,4,9,16],"bp-", lw=5, ms=15, mec="g", mew=5, mfc="r")
plt.show()
```

```python
# 실습 2

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$\pi/2$", r"$\pi$"])
plt.yticks([-1,0,1])
plt.show()
```
